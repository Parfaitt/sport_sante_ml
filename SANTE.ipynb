```python
url="D:\RECO\PROJECT ML\SANTE\dataset_synthetique_5000.csv"
df=pd.read_csv(url)
df.head()
```

```python
df.niveau_activite.unique()
```

```python
#ON REMPLACE EGALEMENT
df['niveau_activite'] = df['niveau_activite'].replace(('Sédentaire', 'Actif','Modéré'),(0,1,2))
df['sexe'] = df['sexe'].replace(('Homme', 'Femme'),(1,0))
```

```python
y = df['poids']
x = df.drop(['poids'], axis = 1)
```

```python
# Train / Test split
X_train, X_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)
model.fit(X_train, y_train)
```

```python
# Performance du modèle

model.score(x, y)
```

```python
import pickle

filename = 'model.sav'
pickle.dump(model, open(filename,'wb',compress=3))
```
